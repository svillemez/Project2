```python
import pandas as pd
```

```python
nyt_file = "all_nyt_books.csv"
nyt_df = pd.read_csv(nyt_file)
```

```python
goodreads_file = "goodreads_ratings_final.csv"
goodreads_df = pd.read_csv(goodreads_file)
```

```python
bookxing_file = "data(jb)/book_data.xlsx"
bookxing_df = pd.read_excel(bookxing_file)
```

```python
nyt_df.head()
```

```python
goodreads_df.head()
```

```python
bookxing_df = bookxing_df.rename(columns={"isbn": "Primary_ISBN_10", "isbn13": "Primary_ISBN_13", "average_rating": "Average_Rating", "ratings_count":"Rating Count", "title":"Title", "authors": "Author"})
bookxing_df.head()
```

```python
nyt_gr_df = nyt_df.merge(goodreads_df, on="Primary_ISBN_13")
```

```python
len(nyt_gr_df)
```

```python
nyt_gr_df.head()
```

```python
nyt_rating_round_list = list(nyt_gr_df["Average_Rating"])
nyt_rating_list = []
```

```python
for rating in nyt_rating_round_list:
    nyt_rating_list.append(int(round(rating,0)))
```

```python
bookx_rating_round_list = list(bookxing_df["Average_Rating"])
bookx_rating_list = []
```

```python
for rating in bookx_rating_round_list:
    bookx_rating_list.append(int(round(rating,0)))
```

```python
bookxing_df["Whole_Rating"] = bookx_rating_list
```

```python
nyt_gr_df["Whole_Rating"] = nyt_rating_list
```
