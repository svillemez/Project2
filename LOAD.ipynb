```python
import pandas as pd
from sqlalchemy import create_engine
from config import postrespw
```

```python
csv_file = "fullnytdata.csv"
nyt_data_df = pd.read_csv(csv_file)
```

```python
nyt_data_df=nyt_data_df.rename(columns={"Rating Count": "Rating_Count"})
nyt_data_df.head()
```

```python
csv_file2 = "fullbookdata.csv"
bxing_data_df = pd.read_csv(csv_file2)
```

```python
bxing_data_df=bxing_data_df.rename(columns={"Rating Count": "Rating_Count"})
bxing_data_df.head()
```

```python
bxing_data_df = bxing_data_df.drop_duplicates(subset="Primary_ISBN_13", keep="first", inplace=False)
```

```python
ny_books = nyt_data_df[["Primary_ISBN_13", "Title", "Author", "Whole_Rating"]]
bx_books = bxing_data_df[["Primary_ISBN_13", "Title", "Author", "Whole_Rating"]]
```

```python
books_data = ny_books.append(bx_books)
books_data
```

```python
ny_ratings = nyt_data_df[["Primary_ISBN_13", "Average_Rating", "Whole_Rating", "Rating_Count"]]
bx_ratings = bxing_data_df[["Primary_ISBN_13", "Average_Rating", "Whole_Rating", "Rating_Count"]]
```

```python
ratings_data = ny_ratings.append(bx_ratings)
ratings_data
```

```python
nyt_data = nyt_data_df[["Primary_ISBN_13", "List", "Weeks_On_List", "Publisher", "Description", "List_Date", "Rank_This_Week","Primary_ISBN_10"]]
nyt_data
```

```python
bookxing_data = bxing_data_df[["Primary_ISBN_13", "book_id", "original_title", "original_publication_year","Primary_ISBN_10"]]
bookxing_data
```

```python
rds_connection_string = (f"postgres:{postrespw}@localhost:5432/ETL_Book_Club")
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
 engine.table_names()
```

```python
bookxing_data.to_sql(name='bookxing_data', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from bookxing_data', con=engine).head()
```

```python
nyt_data.to_sql(name='nyt_data', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from nyt_data', con=engine).head()
```

```python
books_data.to_sql(name='books_data', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from books_data', con=engine).head()
```

```python
ratings_data.to_sql(name='ratings_data', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from ratings_data', con=engine).head()
```
